```python
import shutil
import random
import os
```

```python
train = "isic/train"
test = "isic/test"
val = "isic/val"
```

```python
b_train = 0
b_test = 0
b_val = 0
for image in os.listdir("benign"):
  choice = random.randint(0, 100)
  if choice <= 70:
    shutil.copy(
        os.path.join("benign", image),
        os.path.join(train, "benign")
    )
    b_train += 1
  elif choice <= 85:
    shutil.copy(
        os.path.join("benign", image),
        os.path.join(val, "benign")
    )
    b_val += 1

  else:
     shutil.copy(
        os.path.join("benign", image),
        os.path.join(test, "benign")
    )
     b_test += 1


print(b_train, b_val, b_test)
```

```python
m_train = 0
m_test = 0
m_val = 0
for image in os.listdir("malignant"):
  choice = random.randint(0, 100)
  if choice <= 70:
    shutil.copy(
        os.path.join("malignant", image),
        os.path.join(train, "malignant")
    )
    m_train += 1
  elif choice <= 85:
    shutil.copy(
        os.path.join("malignant", image),
        os.path.join(val, "malignant")
    )
    m_val += 1

  else:
     shutil.copy(
        os.path.join("malignant", image),
        os.path.join(test, "malignant")
    )
     m_test += 1

print(b_train, b_val, b_test)
print(m_train, m_val, m_test)
```
